# 1. Importing Essential Libraries and Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\Users\\Zala Sheshang\\Desktop\\Cognifyz_Internship\\Dataset .csv")
```

# 2. Finding restaurant chains present in the dataset.

```python
# Grouping restaurants by the name
restaurant_chains = df.groupby("Restaurant Name").size().reset_index(name="Chain Count")

# finding the chains of restaurant (frequency > 1)
restaurant_chains = restaurant_chains[restaurant_chains["Chain Count"] > 1]

# finding the top most chains of restaurants
restaurant_chains = restaurant_chains.sort_values(by="Chain Count", ascending=False)

# printing the top restaurant chains
restaurant_chains
```

# 3. Visualise top 10 Restaurant Chains

```python
plt.figure(figsize=(10, 6))
plt.bar(restaurant_chains["Restaurant Name"][:10], restaurant_chains["Chain Count"][:10])
plt.xticks(rotation=40, ha='right')
plt.xlabel("Restaurant")
plt.ylabel("Number of Outlets\n")
plt.title("Top 10 Restaurant Chains")
plt.tight_layout()
plt.show()
```

# 4. Ratings and Popularity of different restaurant chains.

```python
# Average Ratings of different Restaurants group
chains_ratings = df.groupby("Restaurant Name")["Aggregate rating"].mean().reset_index(name="Average Rating")

# Total vots of different Restaurants group
chains_votes = df.groupby("Restaurant Name")["Votes"].sum().reset_index(name="Total Votes")
```

```python
chain_analysis = pd.merge(chains_ratings, chains_votes, on="Restaurant Name")
chain_analysis = chain_analysis.sort_values(by="Average Rating", ascending=False)
chain_analysis
```

# 5. Visualize top 10 Restaurants by Average Rating

```python
plt.figure(figsize=(10, 6))
plt.bar(chain_analysis["Restaurant Name"][:10], chain_analysis["Average Rating"][:10])
plt.xticks(rotation=40, ha='right')
plt.xlabel("Restaurant")
plt.ylabel("Average Rating\n")
plt.title("Top 10 Restaurant Chains by Average Rating")
plt.tight_layout()
plt.show()
```

# 6. Visualize top 10 Restuarants by Total Vots

```python
plt.figure(figsize=(10, 6))
plt.bar(chain_analysis["Restaurant Name"][:10], chain_analysis["Total Votes"][:10])
plt.xticks(rotation=40, ha='right')
plt.xlabel("Restaurant")
plt.ylabel("Total Votes\n")
plt.title("Top 10 Restaurant Chains by Total Votes")
plt.tight_layout()
plt.show()
```
